# 주제 분석(Topic Analysis)
- 텍스트 데이터에 대한 비지도 학습
- 텍스트 데이터에서 **주제(토픽)**를 자동으로 추출하거나 분류하는 자연어 처리(NLP)의 한 기술
- 문서, 리뷰, 뉴스 기사 등 다양한 텍스트 데이터가 가지고 있는 숨겨진 구조를 이해하고자 할 때 사용
- 문서의 단어빈도에 영향을 미치는 잠재 변수가 있다고 가정
> 여기서 잠재변수를 주제(topic) 라고 한다.


## 토픽(Topic)
- 텍스트 데이터에서 반복적으로 나타나는 단어 그룹이나 주제를 말함
- 예:
    - 뉴스 기사에서 "경제", "스포츠", "정치" 등이 주요 토픽일 수 있음
    - 고객 리뷰에서 "서비스 품질", "가격", "상품 기능" 등이 주요 토픽일 수 있음

## 토픽 분석 목적
- 텍스트 데이터의 숨겨진 구조 이해
- 대량의 텍스트를 요약
- 유사한 텍스트 간의 군집화
- 정보 검색등의 시스템 개선

## 토픽 분석의 유형
- 비 지도 학습 기반 토픽 모델링
    - 텍스트 데이터에서 토픽을 자동으로 추출
    - LSA, LDA 등의 방식을 사용
- 지도 학습 기반 토픽 모델링
    - 특정 주제 또는 레이블로 미리 정의된 텍스트를 분류
    - 머신러닝, 딥러닝 모델 등을 사용

## 활용 사례
- 문서 요약
    - 대량의 문서를 요약하여 주요 주제를 파악
- 의견 분석
    - 고객 리뷰에서 긍정적/부정적 요소와 관련된 토픽을 추출
- 추천 시스템
    - 문서나 상품을 유사한 토픽 기반으로 추천
- 뉴스 분류 및 클러스터링
    - 유사한 주제를 가진 뉴스를 자동으로 묶거나 분류
- 소셜 미디어 분석
    - 트윗, 댓글, 포럼 데이터에서 특정 주제의 대화를 분석

# LSA(Latent Semantic Analysis)
- 잠재 의미 분석

- 단어 문서 행렬을 행렬분해하는 방법
    - SVD를 기반으로 한 행렬 분해 방법
    - LSA는 문서-단어 행렬에서 숨겨진 의미를 추출하기 위해 SVD를 사용
- 계산이 비교적 단순하고 통계학에서 많이 사용되는 기법
- 문서들이 단어가 아닌 의미상으로 재배치 되는 효과
- 동음이의어, 오탈자 등일 어느정도 처리가 되는 효과
- 문서에 존재하는 noise가 줄어 전처리 되는 효과

## Truncated SVD 활용한 주제 분석
1. 텍스트 데이터를 벡터화하여 **문서-단어 행렬** 생성
  - 행: 단어
  - 열: 문서
  - 값 (value): 단어의 출현 빈도수

2. SVD 적용하여 행렬 분해
  1. **U (m × k)**:  
     - **문서**와 **주제**(토픽) 간의 관계를 나타내는 행렬입니다.  
    - \( m \): 문서의 개수  
    - \( k \): 주제(토픽)의 개수  
    - **예**: "문서 1은 주제 A에 0.8, 주제 B에 0.2만큼 관련 있다."  

  2. **Σ (k × k)**:  
    - **주제의 중요도**를 나타내는 **대각 행렬**입니다.  
    - 값이 큰 주제일수록 중요한 주제입니다.  
    - **예**: 주제 A가 중요도가 10, 주제 B가 중요도가 5라면 A가 더 중요한 주제입니다.  

  3. **Vᵀ (k × n)**:  
    - **주제**(토픽)와 **단어** 간의 관계를 나타내는 행렬입니다.  
    - \( n \): 단어의 개수  
    - **예**: "단어 '은행'은 주제 A에 0.7, 주제 B에 0.1만큼 관련 있다."  

3. 행렬이 단순화되면 단어와 문서가 의미에 따라 재배치
  - 단어와 문서 간의 관계가 더 잘 드러나게 된다
.

- k : 주제 수
- kxn 은 주제들과 단어들 간의 관련도 값

다운로드.png

1. 원본행렬 A
  - m(문서) * n(단어)
  - 각 단어가 문서에 얼마나 등장했는지 나타냄
2. U 행렬
  - m(문서) * m(문서들을 표현하는 좌표축)
  - 여기서 중요한 정보만 담은 k개의 축만 사용
  - 불필요한 정보 버리기
3. Σ (시그마) 행렬
  - m(문서 수) * n(단어 수)
  - k(특이값의 개수) * k(특이값의 개수)
4. Vᵀ 행렬
  - k(주제의 개수) * n(단어 수)

## CoherenceModel 클래스
- 토픽의 응집력을 평가하는 클래스
- model: 학습된 토픽 모델 객체 전달
  - topics 파라미터에 인수 전달시 생략 가능
- topics: 각 토픽에서 추출된 토큰화된 데이터
- texts: 전체 문서의 토큰화된 데이터
- dictionary: gensim의 딕셔너리객체(어휘집)를 전달
- **점수가 높을수록 단어들이 잘 어울리는 토픽!**

````python
from gensim.models import CoherenceModel
from gensim.corpora import Dictionary
````

````python
dic = Dictionary(tokens_list) # 어휘집
dic
````

````python
# tokens_list : 토큰화된 단어 리스트
# svd : TruncatedSVD 객체
# words : 단어 목록 --- 왜 사용하는지?
# topn : 상위 단어 개수
def coherence_score(tokens_list, svd, words, topn=10):
    # np.argsort(): 배열을 오름차순으로 정렬한 인덱스를 반환
    # [:,::-1] 내림차순으로 변경
    # svd.components_ : Vt 행렬
    # 각 행이 하나의 주제, 각 열은 단어에 대한 가중치(중요도) 나타냄
    tmp = np.argsort(svd.components_, axis=1)[:,::-1] # , 기준으로 행/열
    topics = []
    for arr in tmp[:, :topn]:
        topics.append([ words[i] for i in arr ])

    dic = Dictionary(tokens_list)
    # CoherenceModel 일종의 클래스!
    cos = CoherenceModel(topics=topics, texts=tokens_list, dictionary=dic)
    # get_coherence()는 CoherenceModel 객체가 계산한 주제 응집도 점수를 반환하는 메서드
    return cos.get_coherence() # 응집력 점수 반환
````

````python
# vec = CountVectorizer(max_features = 1000)
# get_feature_names_out() -> 벡터화된 결과의 피처(특징) 이름을 반환하는 메서드
words = vec.get_feature_names_out()
coherence_score(tokens_list,svd,words)
````

주제 응집도 점수
- 0.4 ~ 0.6: **보통(중간 정도)**의 주제 응집도
- 0.6 ~ 0.8: 주제가 의미적으로 잘 구성되어 있으며 품질이 높음

- 응집도 어떻게 높여야 하나?
1. 데이터 전처리 개선
  - 불용어 제거, 단어 정제
  - 자주 등장하는 단어 제거
2. SVD의 n_components 값을 조정하여 주제 수를 늘리거나 줄이기
3. 벡터화 방식 변경
  - CountVectorizer 대신 TfidfVectorizer를 사용하면 단어 가중치를 고려하여 모델링이 개선될 수 있다.
4. 다른 주제 모델 사용
  - LDA (Latent Dirichlet Allocation) 같은 다른거 해보기

````python
scores = []
# k는 주제 수 > 최대 20개
# 주제 수 어떻게 설정하느냐에 따라 응집도 점수 계산
for k in tqdm(range(2,20+1)):
    svd = TruncatedSVD(n_components=k, random_state=SEED)
    svd.fit(x_arr) # x_arr 바탕으로 svd 모델 학습 # fit 메서드는 자기자신(self)를 반환 -> svd
    score = coherence_score(tokens_list, svd, words)
    scores.append(score)
````

````python
fig, ax = plt.subplots()
ax.plot( range(2,21), scores, marker="o" )
ax.set_xticks(range(2,21))
plt.show()
````

# LDA(Latent Dirichlet allocation)
- 잠재 디리클레 할당
- 각 문서에 어떤 주제들이 얼마만큼의 확률로 존재하는지에 대한 확률 모델
- 문서에 잠재된 주제를 추출하여 주제 확률 분포로 나타냄
- 각 문서는 여러 주제의 혼합 분포로 되어 있다고 가정
- 우리가 관찰하는 단어는 주제에서 추출된다고 가정

## LDA의 수행 과정 (by 강사님)
- 깁스 샘플링(Gibbs Sampling) 기법을 사용

1. LDA는 토픽의 개수 k를 입력받는다.
    -  k개의 토픽이 M개의 전체 문서에 걸쳐 분포되어 있다고 가정
    - 예: k=3 → "3개의 주제가 있다"라고 가정
2. LDA는 모든 문서의 모든 단어에 대해서 k개의 토픽 중 하나의 토픽에 랜덤하게 할당한다.
    - 이 작업이 끝나면 각 문서는 토픽 분포를 가지며, 토픽은 단어 분포를 가진 상태
    - 예: 문서1: "강아지", "고양이", "집" → 랜덤하게 주제1, 주제2, 주제3에 각각 할당
3. 모든 문서의 모든 단어에 대해서 k개의 토픽중 하나에 재할당
    - 문서 내의 하나의 단어를 토픽에 재할당 하기 위해 문서내의 토픽 분포와 각 토픽의 단어 분포를 계산
      - 나머지 단어들은 정답인 토픽에 할당되었다고 가정
    - 계산한 분포를 기반으로 각 토픽에 할당될 확률을 계산한 후 가장 높은 확률의 토픽에 재할당
    - 모든 문서의 모든 단어에 대해서 반복
- gensim 의 딕셔너리 객체 만들기
  - 전체문서에 대해 토큰화된 리스트를 전달

## **LDA 수행 과정 이해하기(by 나)**

1. **주제의 개수 \(k\) 설정**  
   - 먼저 사용자가 알고리즘에 **주제의 개수 \(k\)**를 입력합니다.  
     - 예: \(k=3\) → "3개의 주제가 있다"라고 가정합니다.

2. **문서와 주제의 초기 할당**  
   - 모든 문서에 있는 **모든 단어**를 \(k\)개의 주제 중 하나에 **랜덤하게 할당**합니다.  
     - 예:  
       문서1: "강아지", "고양이", "집" → 랜덤하게 주제1, 주제2, 주제3에 각각 할당  
     - 초기에는 아무 정보 없이 무작위로 시작합니다.  

3. **각 문서의 **주제 분포**와 각 주제의 **단어 분포** 계산**  
   - 주제가 단어를 재할당하기 위해 다음 두 가지를 계산합니다:  
     1. **문서 내 주제 분포**:  
        각 문서에 \(k\)개의 주제가 얼마나 할당되어 있는지를 계산합니다.  
        - 예: 문서1 → 주제1(50%), 주제2(30%), 주제3(20%)  
     2. **각 주제의 단어 분포**:  
        각 주제에서 어떤 단어가 자주 등장하는지를 계산합니다.  
        - 예: 주제1 → "강아지"(70%), "고양이"(20%), "집"(10%)

4. **단어 재할당**  
   - 문서 내 각 단어를 한 번씩 다시 주제에 할당합니다.  
   - **나머지 단어들은 이미 정답 주제에 할당되었다고 가정**하고, 특정 단어를 **어떤 주제에 할당할지 확률적으로 계산**합니다.  
   - 가장 확률이 높은 주제에 단어를 재할당합니다.  
   - 예:  
     - 단어 "강아지" → 문서 내 주제 분포와 주제1의 단어 분포를 참고해 **주제1**에 재할당  

5. **반복**  
   - 이 과정을 모든 문서의 모든 단어에 대해 반복합니다.  
   - 몇 번의 반복 후 주제 분포와 단어 분포가 안정되면 LDA 모델이 수렴합니다.


````python
from gensim.corpora import Dictionary # 어휘집
dic = Dictionary(tokens_list)  # tokens_list : 토큰화 된 문서 리스트
````

````python
len(dic) # 전체 어휘 수
````

````python
# 10개 미만의 문서에서 등장한 단어 제거
dic.filter_extremes(no_below=10) # 문서에 거의 등장안하는 단어를 제외해주는 옵션
````

````python
len(dic) # 전체 어휘수 814
````

- corpus 만들기
  - 토큰화된 리스트를 doc2bow(bag-of-words)에 하나의 샘플씩 전달
  - 각 단어를 숫자로 변환하고 단어의 빈도를 함께 저장

````python
corpus = [ dic.doc2bow(tokens) for tokens in tokens_list ]
````

````python
corpus[0]
````

# LdaModel 클래스
- corpus: 샘플별 단어 빈도 데이터
  - 형식 : [(단어ID, 빈도수), (단어ID, 빈도수)]
- num_topics: 토픽수
- id2word: gensim의 딕셔너리 객체
  - 단어 ID와 실제 단어를 매핑
- passes: 반복 횟수(10회 이상할 것)

**LDA 모델을 왜 사용할까?**
- 문서의 숨겨진 주제를 찾을 수 있습니다.
- 각 주제에 대해 어떤 단어가 중요한지 확인할 수 있습니다.
- 주제 모델링은 뉴스 분석, 리뷰 분석, 논문 분석 등에서 널리 활용됩니다.

````python
from gensim.models import LdaModel

# 학습 시작
model = LdaModel(corpus=corpus, num_topics=8, id2word=dic, passes=10, random_state=SEED)
````

- 모델 저장하기

````python
model.save("lda-model")
````

- 저장된 모델 불러오기

````python
model = LdaModel.load("lda-model")
````

- 각 주제의 연관성 높은 단어들 확인해보기

````python
model.print_topics(10)
````

- 문서별로 토픽 확률 보기

````python
# corpus[0] : 첫번째 문서의 Bag-of-Words 벡터
# get_document_topics :문서가 K개의 주제 중 어떤 주제에 속하는지를 확률 분포 형태로 반환

## corpus([0])는 주제 6과 주제 7의 혼합으로 구성되어 있으며:
## 주제 7이 더 높은 확률(55.2%)로 할당되었고,
## 주제 6도 일부(39.4%)를 차지합니다.

model.get_document_topics(corpus[0])
# 제너레이터 같은거 같다
````

- 문서별 토픽 확률 분포를 데이터프레임으로 만들어 보기

````python
df_lda = []
for lst in model.get_document_topics(corpus):
    lst = np.array(lst) # 주제 분포를 NumPy 배열로 변환
    # 주제 수만큼 0으로 채워진 배열 생성 (8개의 주제)
    # 각 주제에 대한 확률 값을 저장하기 위한 초기화된 벡터
    tmp = np.zeros(8)
    idx = lst[:,0].astype(int) # 주제번호를 추출해서 정수형으로 저장
    tmp[idx] = lst[:,1] # 확률값 저장
    df_lda.append(tmp)
````

````python
df_lda = pd.DataFrame(df_lda)
df_lda.head()
````

해석하는 방법
- 토픽번호는 1번 부터 시작
- 왼쪽 그래프
  - 원의 크기가 클수록 토픽 중요도가 높다는 뜻
 - 원 사이의 거리가 클수록 토픽간의 관계가 없다는 뜻
- 오른쪽 그래프
  - 특정 단어 선택시 관련있는 왼쪽그래프의 주제번호에 원의 크기가 커진다.
  - 특정 토픽을 미 선택(Selected Topic == 0)
    - 막대그래프의 x축 방향에 의미는 전체 문서의 단어의 빈도수
    - y축 방향의 단어들들 현저성 순은 나열된다.
      - 현저성: 얼마나 토픽을 구분하는데 눈에 띄게 기여를 했는지를 의미
  - 특정 토픽 선택(Selected Topic == 1 이상 값)
    - 붉은 색 막대그래프의 x축 방향에 의미는 해당 토픽에 단어에 예상되는 빈도수
    - y축방향의 단어들은 해당 토픽에 관련성 순으로 나열된다.

````python
pyLDAvis.display(vis)
# Selected Topic - 토픽번호 1번부터 시작한다 0은 전체를 보겠다는 뜻
# 원이 클수록 중요한 역할을 한다는 뜻
# 원사이가 멀수록 토픽 사이의 관련성이 없다는 뜻 (2,3,4 관련성 높다는 의미)
# 전체로 볼때 오른쪽의 의미 - x 축 방향의 막대그래프 : 빈도수 , y축 순서 : 현저성 의미 (토픽 구분하는데 눈에 띄게 중요한 역할)
# Slide to adjust relevance metric:0.6~0.7 권장
# λ = 1: 단어가 해당 토픽에서 얼마나 자주 등장했는지를 기준으로 단어를 선택.
# λ = 0: 단어가 해당 토픽에만 독특하게 관련이 있는지를 기준으로 단어를 선택.
````

# 주제 다양도
- 각 토픽의 다양성 평가
- 토픽간의 관계를 평가
- 예)
  - 주제수 5개
  - 주제별 관련도가 높은 상위 단어를 10개
  - 주제별 상위 단어들의 합집합(중복제거) / (주제수 X 상위단어 개수)
- 1에 가까울수록 → 주제 간에 중복이 적고 다양성이 높음
- 0에 가까울수록 → 주제 간에 중복된 단어가 많아 다양성이 낮음

````python
topn = 10 # 각 주제에서 상위 10개 단어 사용
s = set() # 중복 제거를 위한 집합 생성
for k in range(8):
    # model.show_topic(k, topn) > [('단어1', 확률1), ('단어2', 확률2), ..., ('단어10', 확률10)]
    # array : numpy 형태로 변환
    arr = np.array(model.show_topic(k, topn))
    # update : 집합 s에 단어 추가
    s.update(arr[:,0]) # 단어만 추출

score = len(s) / ( 8*topn )
score
````
