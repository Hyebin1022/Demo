## 1. MAE 와 RMSE 함수 구현하기
- 다음과 같이 실제 값과 , 모델이 예측한 값이 각각 100개가 있다.
- 실제값과 예측값 사이에 평균 오차를 구하는 함수 두개를 구현해보자.

````python
np.random.seed(42)
target = np.random.randint(50, 91,100) # 정답값
pred = np.random.randint(40, 100,100) # 예측값
target , pred
````

### MAE(Mean Absolute Error)
- 실제 값과 예측 값의 차이를 절대값으로 변환해 평균
$$
MAE = \frac{1}{n}{\sum_{i=1}^{n}|y_i-\hat{y}_i|}
$$

```
아래 셀의 코드 작성하여 함수구현
```

````python
def mae(target,pred):
    return np.mean(abs(np.subtract(target,pred)))
mae(target,pred)
````

### RMSE(Root Mean Squared Error)
- 실제 값과 예측 값의 차이를 제곱한 뒤 평균한 후에 루트

$$
RMSE = \sqrt{\frac{1}{n}{\sum_{i=1}^{n}(y_i-\hat{y}_i)^2}}
$$

```
아래 셀의 코드 작성하여 함수구현
```

````python
def rmse(target,pred):
    return (np.mean(np.square(target-pred)))**0.5

rmse(target,pred)
````

* 11번은 반복문 써야함

````python
################ 제너레이터 
def split_dataset(data):
    in_dex = np.arange(len(data))
    np.random.shuffle(in_dex)
    one_five_len = int(np.ceil(len(data) * 0.2)) 
    for i in range(5):
        start = i * one_five_len
        end = (i + 1) * one_five_len
        yield in_dex[start:end]  # 인덱스 반환

gen = split_dataset(x_train)
for idx in gen:
    print(idx)
````

* yield from 사용한 방식

````python
def split_dataset(dataset):
    n_split = 5
    n_row=len(dataset)
    index_arr = np.arange(n_row)
    n_samples = n_row//n_split
    np.random.shuffle(index_arr)
    # 나머지체크
    n_sub = n_row - n_samples *n_split
    index_list = index_arr[:n_row-n_sub].reshape(n_split,n_samples).tolist()

    if n_sub > 0 :
        index_list[-1].extend(index_arr[n_row-n_sub:])

    yield from index_list
split_dataset(x_train)
````

### 시계열 학습데이터 형태로 반환하는 함수를 만드시오.
- samsung_stock 변수의 데이터는 2차원 데이터이다.
- 학습데이터는 행별로 10일치의 행과 4열(시작가 ,상한가 , 하한가, 종가) 형태의 행렬을 담고 있는 3차원 형태의 데이터셋으로 변경하시오.
- 정답데이터는 미래의 5일치의 종가를 예측해야 한다는 가정하에 행별로 미래의 5일치의 종가가 담긴 2차원 형태의 데이터셋으로 변경하시오.
- 1일씩 이동하면서 데이터셋을 구성하시오.

```
아래 셀의 코드 작성하여 함수 구현
```

````python
def transform_data(samsung_stock,seq_len=10,pred_len=5):
    x_train = []
    y_train = []
    len_stock = len(samsung_stock) # 243

    for i in range(len_stock - seq_len - pred_len + 1):
        x_train.append(samsung_stock[i:i + seq_len])
        y_train.append(samsung_stock[i + seq_len :i + seq_len + pred_len,-1])
    
    return np.array(x_train), np.array(y_train)
````

-------------------------------------

- 다음 셀들은 위에 함수 실행 결과를 검증하는 부분

````python
x_train , y_train = transform_data(samsung_stock)
x_train.shape, y_train.shape # ((234, 10, 4), (234, 5))
````
